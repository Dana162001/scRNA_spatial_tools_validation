```python
# Group the cells by spatial cluster and cell type, then count the number of cells in each group
cell_counts = adata.obs.groupby(['spatial_cluster_15', 'tacco']).size().reset_index(name='cell_count')

# Plot a barplot for each spatial cluster
for spatial_cluster in adata.obs['spatial_cluster_15'].unique():
    # Filter the data for the current spatial cluster
    cluster_data = cell_counts[cell_counts['spatial_cluster_15'] == spatial_cluster]
    
    # Create the barplot
    plt.figure(figsize=(10, 6))
    sns.barplot(x='tacco', y='cell_count', data=cluster_data)
    plt.title(f'Cell Type Distribution in Spatial Cluster {spatial_cluster}')
    plt.xlabel('Cell Type')
    plt.ylabel('Cell Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
cell_counts = adata.obs.groupby(['spatial_cluster_15', 'tacco']).size().reset_index(name='cell_count')
pivot_table = cell_counts.pivot_table(index='spatial_cluster_15', columns='tacco', values='cell_count', fill_value=0)

plt.figure(figsize=(100, 70))
pivot_table.plot(kind='bar', stacked=True, colormap=cc.cm.glasbey_bw)
plt.title('Cell Type Distribution in Spatial Clusters')
plt.xlabel('Spatial Cluster')
plt.ylabel('Cell Count')
plt.xticks(rotation=45, ha='right') 
plt.legend(title='Cell Type', bbox_to_anchor=(1,1), loc='upper left', fontsize='xx-small')
plt.tight_layout()
plt.show()
```

```python
cell_counts = adata.obs.groupby(['spatial_cluster_15', 'tacco']).size().reset_index(name='cell_count')
pivot_table = cell_counts.pivot_table(index='spatial_cluster_15', columns='tacco', values='cell_count', fill_value=0)
cluster_total = pivot_table.sum(axis=1)
percentage_table = pivot_table.div(cluster_total, axis=0) * 100

plt.figure(figsize=(12, 8))
percentage_table.plot(kind='bar', stacked=True, colormap=cc.cm.glasbey_bw)
plt.title('Cell Type Distribution in Spatial Clusters')
plt.xlabel('Spatial Cluster')
plt.ylabel('Percentage of Cells')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Cell Type', bbox_to_anchor=(1.05, 1), loc='upper left',fontsize='xx-small')
plt.tight_layout()
plt.show()
```

```python
# Get the top 10 genes for each spatial cluster
top_genes_by_cluster = {}

# Iterate over each spatial cluster
for cluster in adata.obs['spatial_cluster_5'].unique():
    # Get the gene expression data for the current cluster
    cluster_data = adata[adata.obs['spatial_cluster_5'] == cluster]
    # Get the mean expression of each gene across cells in the cluster
    mean_expression = cluster_data.X.mean(axis=0)
    # Get the indices of the top 10 genes by mean expression
    top_gene_indices = np.argsort(mean_expression)[::-1][:10]
    # Get the gene names corresponding to the top gene indices
    top_genes = cluster_data.var_names[top_gene_indices].tolist()
    # Store the top genes for the current cluster
    top_genes_by_cluster[cluster] = top_genes

# Print the top 10 genes for each spatial cluster
for cluster, genes in top_genes_by_cluster.items():
    print(f"Spatial Cluster {cluster}:")
    print(genes)
    print()
```

```python
def get_top_genes_by_cluster(andata, cluster_col_name):
    top_genes_by_cluster = {}


    for cluster in andata.obs[cluster_col_name].unique():
        cluster_data = andata[andata.obs[cluster_col_name] == cluster]
        mean_expression = cluster_data.X.mean(axis=0)
        top_gene_indices = np.argsort(mean_expression)[::-1][:10]
        top_genes = cluster_data.var_names[top_gene_indices].tolist()
        top_genes_by_cluster[cluster] = top_genes
    return top_genes_by_cluster
```

```python
top_genes = get_top_genes_by_cluster(adata, 'spatial_cluster_10')
```
